```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'survey_final.csv'
read = pd.read_csv(file)
```

```python
df = pd.DataFrame(read)
```

```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
df = df.drop(columns = {'Hobbyist', 'DatabaseDesireNextYear', 'Year', 'LanguageDesireNextYear', 'OrgSize'})
```

```python
df['ConvertedComp'] = round(df['ConvertedComp'],2)
```

```python
df = df.rename(columns = {'ConvertedComp': 'Salary', 'JobSat': 'Job Satisfaction', 'DatabaseWorkedWith': 'Database Use', 'LanguageWorkedWith': 'Programming Language', 'UndergradMajor': 'Major', 'YearsCodePro': 'Years Coding Professionally', 'EdLevel': 'Education Level', 'DevType': 'Developer'})
```

```python
df
```

```python
new_df = df.groupby(df['Country'])

mean_df = new_df.mean()
mean_df['Salary'] = round(mean_df['Salary'],2)
mean_df
```

```python
new_labels = []
new_labels = mean_df.index
```

```python
dims = (45, 10)
fig, ax = plt.subplots(figsize=dims)
chart = sns.barplot(x=mean_df.index, y=mean_df['Salary'])
plt.xticks(rotation=90)
```
